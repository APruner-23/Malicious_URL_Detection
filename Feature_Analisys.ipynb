```python
# import a file from a github repo here is the raw link : https://raw.githubusercontent.com/APruner-23/Malicious_URL_Detection/main/updated_malicious_phish_with_tld.csv

import pandas as pd

dataset_url = "https://raw.githubusercontent.com/APruner-23/Malicious_URL_Detection/main/updated_malicious_phish_with_tld.csv"

# Read the CSV file from the URL
df = pd.read_csv(dataset_url)

# Print the DataFrame
print(df.head())
```

**evaluate the importance of attributes in a dataset using SelectKBest utility from sklearn**

```python
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
X = df.iloc[:, 2:]  # Feature matrix
y = df.iloc[:, 1]   # Target clasification

# Select features based on correlation
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, y)

# Get scores for all features
scores = selector.scores_
p_values = selector.pvalues_

# Combine features and scores into a DataFrame for better readability
feature_scores = pd.DataFrame({
    'Feature': X.columns,
    'Score': scores,
    'p-value': p_values
})

# Sort features by score

feature_scores.sort_values(by='Score', ascending=False, inplace=True)
feature_scores.drop(columns=['p-value'], inplace=True)
print("top 5 Features sorted by importance and correlation to the target labels")
print(feature_scores.head())
```
